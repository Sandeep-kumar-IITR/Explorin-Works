```python
import pandas as pd
import seaborn as sns
vil_data=pd.read_csv("village_data.csv")
vil_data.head(1).dtypes
```

```python
# goverment higest education
vill_data_gov_art = vil_data.groupby("govt_art_science_degree_college").get_group(0)

vill_data_gov_eng = vil_data.groupby("govt_engineering_college").get_group(0)

vill_data_gov_med = vil_data.groupby("govt_medicine_college").get_group(0)

vill_data_gov_man = vil_data.groupby("govt_management_college").get_group(0)

vill_data_gov_poly = vil_data.groupby("govt_polytechnic_college").get_group(0)

vill_data_gov_iti = vil_data.groupby("govt_vocational_ITI_training").get_group(0)

vill_data_gov_tra = vil_data.groupby("govt_non_formal_training_centre").get_group(0)

vill_data_gov_dis = vil_data.groupby("govt_disabled_school").get_group(0)

vill_data_gov_oth = vil_data.groupby("govt_other_education").get_group(0)


vill_data_gov_all = pd.concat([vill_data_gov_art,vill_data_gov_eng,vill_data_gov_med,vill_data_gov_man,vill_data_gov_poly,vill_data_gov_iti,vill_data_gov_tra,vill_data_gov_dis,vill_data_gov_oth])

vill_data_gov=vill_data_gov_all.drop_duplicates()
vill_data_gov
```

```python
# pvt higest education
vill_data_pvt_art = vil_data.groupby("pvt_art_science_degree_college").get_group(0)

vill_data_pvt_eng = vil_data.groupby("pvt_engineering_college").get_group(0)

vill_data_pvt_med = vil_data.groupby("pvt_medicine_college").get_group(0)

vill_data_pvt_man = vil_data.groupby("pvt_management_college").get_group(0)

vill_data_pvt_poly = vil_data.groupby("pvt_polytechnic_college").get_group(0)

vill_data_pvt_iti = vil_data.groupby("pvt_vocational_ITI_training").get_group(0)

vill_data_pvt_tra = vil_data.groupby("pvt_non_formal_training_centre").get_group(0)

vill_data_pvt_dis = vil_data.groupby("pvt_disabled_school").get_group(0)

vill_data_pvt_oth = vil_data.groupby("pvt_other_education").get_group(0)


vill_data_pvt_all = pd.concat([vill_data_pvt_art,vill_data_pvt_eng,vill_data_pvt_med,vill_data_pvt_man,vill_data_pvt_poly,vill_data_pvt_iti,vill_data_pvt_tra,vill_data_pvt_dis,vill_data_pvt_oth])

vill_data_pvt_code=vill_data_pvt_all["district_code"].nunique()
print(vill_data_pvt_code)
vill_data_pvt = vill_data_pvt_all.drop_duplicates()
vill_data_pvt
```

```python
#total highest education

vill_data_high_all = pd.merge(vill_data_gov, vill_data_pvt, how='inner')
vill_data_high_all

vill_data_high_all['total_schools'] = vill_data_high_all['govt_pre_primary_school_UKG']+ vill_data_high_all['pvt_pre_primary_school_UKG']+vill_data_high_all['govt_primary_school']+ vill_data_high_all['pvt_primary_school']+vill_data_high_all['govt_secondary_school']+ vill_data_high_all['pvt_secondary_school']+vill_data_high_all['govt_senior_secondary_school']+ vill_data_high_all['pvt_senior_secondary_school']+vill_data_high_all['govt_middle_school']+ vill_data_high_all['pvt_middle_school']
vill_data_high_all
# ind = vill_data_gov.district_code.isin(vill_data_pvt.district_code) & vill_data_pvt.district_code.isin(vill_data_gov.district_code)
# ind
# #vill_data_all = vill_data_gov[ind]
# #vill_data_all.append(vill_data_pvt[ind])



# # vill_total_high = pd.merge(vill_data_gov,vill_data_pvt,how = 'inner' ,on = ["district_code"])
# # # vill_total_high= vill_total_high_all.drop_duplicates()
# # vill_total_high
```
